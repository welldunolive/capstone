```python
ddf = pd.read_csv('data/steam_games.csv')
```

```python
ddf.head()
```

```python
ddf.shape
```

```python
ddf_new = ddf.drop(columns = ['types', 'desc_snippet', 'languages', 'achievements', 'game_details', 'game_description', 'mature_content', 'minimum_requirements', 'recommended_requirements', 'discount_price'])
```

```python
ddf_new.shape
```

```python
ddf_new = ddf_new.apply(lambda x: x.astype(str).str.lower())
```

```python
ddf_new.head()
```

```python
ddf_new['recent_reviews'].unique()
```

```python
ddf_new.loc[ddf_new['recent_reviews'].str.startswith('very positive'), 'recent_reviews'] = 'very positive'

ddf_new.loc[ddf_new['recent_reviews'].str.startswith('mixed'), 'recent_reviews'] = 'mixed'

ddf_new.loc[ddf_new['recent_reviews'].str.startswith('mostly positive'), 'recent_reviews'] = 'mostly positive'

ddf_new.loc[ddf_new['recent_reviews'].str.startswith('overwhelmingly positive'), 'recent_reviews'] = 'overwhelmingly positive'

ddf_new.loc[ddf_new['recent_reviews'].str.startswith('very negative'), 'recent_reviews'] = 'very negative'

ddf_new.loc[ddf_new['recent_reviews'].str.startswith('mostly negative'), 'recent_reviews'] = 'mostly negative'

ddf_new.loc[ddf_new['recent_reviews'].str.startswith('overwhelmingly negative'), 'recent_reviews'] = 'overwhelmingly negative'
```

```python
ddf_new['recent_reviews'].unique()
```

```python
ddf_new.loc[ddf_new['all_reviews'].str.startswith('very positive'), 'all_reviews'] = 'very positive'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('mixed'), 'all_reviews'] = 'mixed'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('mostly positive'), 'all_reviews'] = 'mostly positive'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('overwhelmingly positive'), 'all_reviews'] = 'overwhelmingly positive'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('very negative'), 'all_reviews'] = 'very negative'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('mostly negative'), 'all_reviews'] = 'mostly negative'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('overwhelmingly negative'), 'all_reviews'] = 'overwhelmingly negative'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('positive'), 'all_reviews'] = 'positive'

ddf_new.loc[ddf_new['all_reviews'].str.startswith('negative'), 'all_reviews'] = 'negative'
```

```python
ddf_new['all_reviews'].unique()
```

```python
ddf_new.loc[ddf_new['all_reviews'].str.endswith('need more user reviews to generate a score'),  'all_reviews'] = 'nan'

ddf_new.loc[ddf_new['all_reviews'].str.endswith('need more user reviews to generate a score'),  'all_reviews'] = 'nan'
```

```python
ddf_new['all_reviews'].unique()
```

```python
ddf_new['release_date'] = [x[-4:] for x in ddf_new['release_date']]
```

```python
ddf_new
```

```python
ddf_new['release_date'] = ddf_new['release_date'].apply(lambda x: np.where(x.isdigit(), x, '1900'))
```

```python
ddf_new['release_date']
```

```python
ddf_new['release_date'].value_counts()
```

```python
ddf_new.loc[ddf_new['release_date'].str.endswith('1201'),  'release_date'] = '1900'
```

```python
ddf_new.loc[ddf_new['release_date'].str.endswith('0'),  'release_date'] = '1900'
```

```python
ddf_new['release_date'].value_counts()
```
