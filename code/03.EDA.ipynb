```python
from wordcloud import WordCloud, STOPWORDS
```

```python
# Stopwords, https://www.geeksforgeeks.org/generating-word-cloud-python/

comment_words = '' 
stopwords = set(STOPWORDS) 
  
# iterate through the csv file 
for val in df['genre']: 
      
    # typecaste each val to string 
    val = str(val) 
  
    # split the value 
    tokens = val.split() 
      
    # Converts each token into lowercase 
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
      
    comment_words += " ".join(tokens)+" "
  
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stopwords, 
                min_font_size = 10).generate(comment_words) 
  
# plot the WordCloud image                        
plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 
  
plt.show()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cvect = CountVectorizer(stop_words = "english")
```

```python
df['genre'].dropna(inplace = True)
```

```python
words = cvect.fit_transform(df['genre'])
```

```python
df_words = pd.DataFrame(words.toarray(), columns=cvect.get_feature_names())
```

```python
#https://medium.com/@cristhianboujon/how-to-list-the-most-common-words-from-text-corpus-using-scikit-learn-dad4d0cab41d

sum_words = words.sum(axis=0)
words_freq = [(word, sum_words[0, idx]) for word, idx in cvect.vocabulary_.items()]     
words_freq = sorted(words_freq, key = lambda x: x[1], reverse=True)
words_freq[:25]
```

```python
word_list = []
freq_list = []
i = 0
for word in words_freq:
    word_list.append(word[0])
    freq_list.append(word[1])
    i += 1
    if i > 5:
        break
word_list
```

```python
plt.figure(figsize=(8, 8))
plt.barh(word_list, freq_list)
plt.xlabel('Frequency')
plt.title('Top 6 Genres', fontsize=20)
```
