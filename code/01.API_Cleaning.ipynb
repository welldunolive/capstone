```python
import pandas as pd
```

```python
action = pd.read_csv('data/action.csv')
adventure = pd.read_csv('data/adventure.csv')
early= pd.read_csv('data/early.csv')
ex = pd.read_csv('data/ex.csv')
free = pd.read_csv('data/free.csv')
indie = pd.read_csv('data/indie.csv')
mmo = pd.read_csv('data/mmo.csv')
rpg = pd.read_csv('data/rpg.csv')
simulation = pd.read_csv('data/simulation.csv')
sports = pd.read_csv('data/sports.csv')
strategy = pd.read_csv('data/strategy.csv')
```
