```python
import pandas as pd
df = pd.read_csv('../Data/final_df.csv')
```

```python
duplicats
```

```python
import numpy as np

images = []
for i in df['url']:
    r = requests.get(i)
    if r.status_code ==200:
        soup = BeautifulSoup(r.text, 'lxml')
        pic = soup.find('img', {'class': 'game_header_image_full'})
        if pic is None:
            images.append(np.nan)
        else:
            images.append(pic.attrs['src'])
    else:
        images.append(np.nan)
```

```python
images
```

```python
df['images'] = images
```

```python
df.head()
```

```python
duplicates = df[df.duplicated('url')]
```

```python
duplicates
```

```python
df.drop_duplicates(subset = ['url'], keep = 'last', inplace = True)
```
